```python
# Import libraries
import pandas as pd
import os
import shutil
import random
```

```python
# Define paths to this dataset (https://github.com/ieee8023/covid-chestxray-dataset) 
# which I already downloaded it to my computer.
FILE_PATH = "covid-chestxray-dataset/metadata.csv"
IMAGE_PATH = "covid-chestxray-dataset/images"
```

```python
# Read csv file
df = pd.read_csv(FILE_PATH)
df.head()
```

```python
# Create needed folders
os.mkdir("Dataset/Positive")
os.mkdir("Dataset/Negative")

os.mkdir("Dataset/Train")
os.mkdir("Dataset/Train/Positive")
os.mkdir("Dataset/Train/Negative")

os.mkdir("Dataset/Test")
os.mkdir("Dataset/Test/Positive")
os.mkdir("Dataset/Test/Negative")
```

```python
# Copy x-ray images of all COVID-19 cases from this dataset
target_dir = "Dataset/Positive"

n_covid = 0

for i, row in df.iterrows():
    if row['finding'] == 'Pneumonia/Viral/COVID-19':
        filename = row['filename']
        image_path = os.path.join(IMAGE_PATH, filename) # image that want to copy
        image_copy_path = os.path.join(target_dir, filename) # copy to
        if os.path.exists(image_path):
            shutil.copy2(image_path, image_copy_path)
            n_covid += 1
        
print("{} COVID cases".format(n_covid))
```

```python
# Define paths to this dataset (https://www.kaggle.com/andrewmvd/convid19-x-rays?select=xrays.csv.) 
# which I already downloaded it to my computer.
FILE_PATH2 = "COVID-19-XRay/xrays.csv"
IMAGE_PATH2 = "COVID-19-XRay/X-rays"
```

```python
# Read the cvs file and view
df2 = pd.read_csv(FILE_PATH2)
df2.head()
```

```python
# Copy x-ray image of all COVID-19 cases
n_covid2 = 0

for i, row in df2.iterrows():
    if row['COVID-19'] == 1:
        ID = str(row['ID'])
        filename = ID + '.jpeg'
        image_path = os.path.join(IMAGE_PATH2, filename) # image that want to copy
        image_copy_path = os.path.join(target_dir, filename) # copy to
        if os.path.exists(image_path):
            shutil.copy2(image_path, image_copy_path)
            n_covid2 += 1
        else:
            filename = ID + '.jpg'
            image_path = os.path.join(IMAGE_PATH2, filename) # image that want to copy
            image_copy_path = os.path.join(target_dir, filename) # copy to
            if os.path.exists(image_path):
                shutil.copy2(image_path, image_copy_path)
                n_covid2 += 1
            
        
print("{} COVID cases".format(n_covid2))
```

```python
# Negative cases
# Copy x-ray images of other types of Pneumonia except 'No Finding', 'Unknown' and 'todo'

target_dir = "Dataset/Negative"

n = 0
exclude = ['Pneumonia/Viral/COVID-19', 'No Finding', 'Unknown', 'todo']

for i, row in df.iterrows():
    if not row['finding'] in exclude:
        filename = row['filename']
        image_path = os.path.join(IMAGE_PATH, filename) # image that want to copy
        image_copy_path = os.path.join(target_dir, filename) # copy to
        if os.path.exists(image_path):
            shutil.copy2(image_path, image_copy_path)
            n += 1
        
print("{} Non-COVID cases".format(n))
```

```python
# Define paths to this dataset (https://www.kaggle.com/paultimothymooney/chest-xray-pneumonia) 
# which I already downloaded it to my computer.
IMAGE_PATH3 = "chest_xray/train/PNEUMONIA"
image_names = os.listdir(IMAGE_PATH3)
random.shuffle(image_names)
```

```python
n_more = n_covid + n_covid2 - n

for i in range(n_more):
    filename = image_names[i]
    image_path = os.path.join(IMAGE_PATH3, filename) # image that want to copy
    image_copy_path = os.path.join(target_dir, filename) # copy to
    shutil.copy2(image_path, image_copy_path)
    
print("{} Non-COVID cases".format(n_more))
```

```python
n_covid += n_covid2
n_non_covid = n + n_more
print("{} COVID cases.".format(n_covid))
print("{} Non COVID cases.".format(n_non_covid))
```

# Make training and testing sets

```python
POS_PATH = "Dataset/Positive"
NEG_PATH = "Dataset/Negative"
pos_imgs = os.listdir("Dataset/Positive") # filenames of picture of positive cases
neg_imgs = os.listdir("Dataset/Negative") # filenames of picture of negative cases

# Shuffle filenames
random.shuffle(pos_imgs)
random.shuffle(neg_imgs)
```

```python
# 80% training set, 20% testing set
n_test = int(len(pos_imgs) * 0.2)
n_train = len(pos_imgs) - n_test
print("Test set has {} images".format(n_test))
print("Train set has {} images".format(n_train))
```

```python
# Test posive
target_dir = "Dataset/Test/Positive"

for i in range(n_test):
    filename = pos_imgs[i]
    image_path = os.path.join(POS_PATH, filename) # image that want to copy
    image_copy_path = os.path.join(target_dir, filename) # copy to
    shutil.copy2(image_path, image_copy_path)
    print('Moving image {} to test positive'.format(i))
```

```python
# Train positive
target_dir = "Dataset/Train/Positive"
for i in range(n_test, len(pos_imgs)):
    filename = pos_imgs[i]
    image_path = os.path.join(POS_PATH, filename) # image that want to copy
    image_copy_path = os.path.join(target_dir, filename) # copy to
    shutil.copy2(image_path, image_copy_path)
    print('Moving image {} to train positive'.format(i))
```

```python
# Train negative
target_dir = "Dataset/Train/Negative"

for i in range(n_test, len(neg_imgs)):
    filename = neg_imgs[i]
    image_path = os.path.join(NEG_PATH, filename) # image that want to copy
    image_copy_path = os.path.join(target_dir, filename) # copy to
    shutil.copy2(image_path, image_copy_path)
    print('Moving image {} to train negative'.format(i))
```

```python
# Test negative
target_dir = "Dataset/Test/Negative"

for i in range(n_test):
    filename = neg_imgs[i]
    image_path = os.path.join(NEG_PATH, filename) # image that want to copy
    image_copy_path = os.path.join(target_dir, filename) # copy to
    shutil.copy2(image_path, image_copy_path)
    print('Moving image {} to test negative'.format(i))
```
